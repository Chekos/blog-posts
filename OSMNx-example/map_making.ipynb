```python
import osmnx as ox
%matplotlib inline
```

```python
address = 'Lindavista, Tijuana, Municipio de Tijuana, Baja California, Mexico'
distance_in_meters = 5500
distance_type = 'network'  # this is either 'network' or 'bbox' explore what each mean on the docs

tijuana = ox.graph_from_address(address, 
                                distance = distance_in_meters, 
                                distance_type = distance_type,
                               )
```

***
This is the default look for our maps.

```python
ox.plot_graph(tijuana,
              save = True,
              filename = 'tijuana_draft'
             );
```

*** 
But they are easily customizable.

We will change:
1. `fig_height` to 8 inches
2. `bgcolor` or background color to `#F0F0E0` a cream color
3. `node_alpha` or the transparency of nodes to 0 (we just don't want nodes in our map really)
4. `edge_color` or the color of the 'edges' or what would be streets in our map to `#3E3E3E` (basically black but not really)
5. `dpi` or 'dots per inch' aka the resolution to 600. The default is 300 and it's already pretty good.
6. `save` to `True`
7. `filename` equal to whatever you want to name this figure

You can also change the `file_format` (the default is `.png` but it can be `jpg` or `svg`

```python
ox.plot_graph(G = tijuana, 
              fig_height = 8,
              bgcolor="#f0f0e0", 
              node_alpha=0, 
              edge_color="#3e3e3e",
              dpi = 600,
              save = True,
              filename = 'tijuana',
              
             );
```
